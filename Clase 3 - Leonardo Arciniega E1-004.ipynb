##I. Introducción a la programación:

###1. ¿Qué es un programa?

Un programa es el alma de la computadora, contiene las órdenes que ejecuta el ordenador para alcanzar un objetivo determinado.

Las computadoras por sí mismas no son capaces de hacer nada, necesitan programas que les indiquen qué hacer y cómo hacerlo, por ello un programa tiene que ser secuencial y específico. Estos programas están dados en un lenguaje de programación.

###2. Lenguaje de Programación:

Son secuencias estructuradas y precisas, que conforman un conjunto de instrucciones o IL (Instruction List).

####Composición de un Lenguaje de Programación:
#####a) Alfabeto:
En Python, el alfabeto está compuesto por letras minúsculas y mayúsculas del alfabeto latino, números, y una serie de caracteres especiales.

#####b) Léxico:
Son comandos específicos para realizar una instrucción determinada.

#####c) Sintaxis:
Se refiere a la estructura que debe respetar el código para formar un comando válido.

#####d) Semántica:
Determina que un código tenga sentido y una función lógica.

###3. Tipos de lenguaje:

####a) Lenguaje máquina:
También conocido como IL, es la serie binaria (unos y ceros) que entiende una computadora a la hora de ejecutar un programa.

####b) Lenguaje de Alto Nivel:
Es el lenguaje que funciona de puente entre la máquina y el programador, pues aunque es más complejo que los códigos binarios de la máquina, tiene una sintaxis entendible para el humano.

####c) Código fuente:
Es el conjunto de instrucciones escrito en un lenguaje de programación de alto nivel específico.

###4. Ejecución del código:

Es importante que el código respete la composición de un lenguaje de programación antes descrita, caso contrario el programa que ejecuta el código enviará alertas por los errores en el mismo, y no podrá ejecutarlo.

## II. Introducción a Python

### 1. ¿Qué es Python?
Python, desarrollado por Guido van Rossum, es un lenguaje de programación de alto nivel, interpretado, orientado a objetos y de propósito general con semántica dinámica. Se utiliza para una amplia variedad de aplicaciones de programación.

#### a) Origen de Python
El nombre "Python" proviene de la serie de comedia de la BBC "Monty Python's Flying Circus", y no de la serpiente. Durante el auge de la serie, el equipo de Monty Python actuaba en vivo en lugares como el Hollywood Bowl. Guido van Rossum, siendo un fanático del show, decidió nombrar al lenguaje en su honor.

### 2. ¿Por qué aprender Python?
Python es altamente versátil y poderoso, presentando múltiples beneficios:

- Su sintaxis es simple y legible, ideal para principiantes.
- Se aplica en diversas áreas, desde el desarrollo web hasta la inteligencia artificial.
- Es un software de código abierto con una comunidad robusta que aporta bibliotecas y herramientas.
- Tiene alta demanda en el mercado laboral, con numerosas oportunidades para programadores de Python.

### 3. Características de Python:
- Es un lenguaje interpretado.
- Posee una sintaxis clara y fácil de leer.
- Admite múltiples paradigmas: procedural, orientado a objetos y funcional.
- Cuenta con una extensa biblioteca estándar.

## III. Cardinalidad
En matemáticas, la cardinalidad de un conjunto es la medida del "número de elementos" en el conjunto. Por ejemplo, el conjunto A = {2, 4, 6} tiene tres elementos, por lo tanto, su cardinalidad es 3. La cardinalidad se puede abordar comparando conjuntos mediante biyecciones e inyecciones, o utilizando números cardinales.

Usualmente, la cardinalidad de un conjunto A se denota como |A|, similar a la notación de valor absoluto, aunque su significado depende del contexto. Alternativamente, se puede representar como n(A), A, card(A) o #A.

### 1. Conjuntos Cardinales

- |A| = |B|: Dos conjuntos A y B tienen la misma cardinalidad si existe una biyección (función inyectiva y suprayectiva) entre ellos. Esto se denota como A ≈ B o w ~ B. Por ejemplo, el conjunto E = {0, 2, 4, 6, ...} de números pares no negativos tiene la misma cardinalidad que el conjunto N = {0, 1, 2, 3, ...} de números naturales, ya que la función f(n) = 2n es una biyección de N sobre E.
- |A| ≤ |B|: A tiene una cardinalidad menor o igual a B si existe una función inyectiva de A en B.
- |A| < |B|: A tiene una cardinalidad estrictamente menor a B si existe una función inyectiva pero no biyectiva de A en B.

Ejemplo:
El conjunto N de números naturales tiene una cardinalidad menor que el conjunto R de números reales, ya que la función inclusión i : N → R es inyectiva, pero no existe una función biyectiva de N en R.

El teorema de Cantor-Bernstein-Schroeder establece que si |A| ≤ |B| y |B| ≤ |A|, entonces |A| = |B|. Los axiomas de un preorden también son validados por ≤. El axioma de elección afirma que dicho preorden es total, es decir, |A| ≤ |B| o |B| ≤ |A| para todo A y B.

### 2. Evolución de los Sistemas Numéricos
Desde los números naturales hasta conjuntos más complejos, los sistemas numéricos han evolucionado. A partir de los números naturales, se han desarrollado sistemas más sofisticados:

| Sistema Numérico | Ejemplo de Problema | Solución Propuesta |
| ---------------- | --------------------- | -------------------- |
| Números Naturales | N/A | Base para otros sistemas numéricos |
| Números Enteros | -1 y +1 | Adición de números negativos |
| Números Racionales | $1/2$ y $2/3$ | Fracciones para expresar relaciones |
| Números Reales | $√2$ | Inclusión de números irracionales |
| Números Complejos | $√(-1) = i$ | Extensión a un plano complejo con partes reales e imaginarias |

### 3. Números Complejos
Los números complejos consisten en una parte real y una parte imaginaria, representándose en un plano complejo. Un número complejo (a + bi) tiene su conjugado (a - bi), que es simétrico respecto al eje real en el plano complejo.

#### a) Representación de Números Complejos
- Número Complejo: (a + bi)
- Conjugado: (a - bi)

### 4. Conjunto de Subconjuntos: Conjunto Potencial
El conjunto potencial de un conjunto Ω es el conjunto de todos los subconjuntos posibles de Ω.

| Conjunto Original (Ω) | Subconjuntos | Cardinal del Conjunto Potencial |
| --------------------- | ------------- | ------------------------------ |
| {1, 2, 3} | {∅, {1}, {2}, {3}, {1,2}, {1,3}, {2,3}, {1,2,3}} | 8 (2^3) |

El cardinal del conjunto potencial de un conjunto finito es 2^n, donde n es el número de elementos en el conjunto original.

### 5. Producto Cartesiano
El producto cartesiano de dos conjuntos A y B es el conjunto de todos los pares ordenados [a, b], donde a ∈ A y b ∈ B.

Ejemplo:
Para A = {1, 2, 3} y B = {0, 1}:

#### a) Producto Cartesiano (A × B)
{(1,0), (1,1), (2,0), (2,1), (3,0), (3,1)}

El cardinal de (A × B) es (|A| × |B|).

### 6. Espacios de Dimensiones Superiores
Desde $ℝ^2$, se pueden extender a dimensiones superiores como $ℝ^n$, que representa un espacio de n-tuplas de números reales.

#### a) Ejemplo de Notación
- $ℝ^2$ es el plano cartesiano.
- $ℝ^3$ es el espacio tridimensional.
- $ℝ^n$ es el conjunto de todas las n-tuplas de números reales.

### 7. Cardinalidad de Conjuntos Infinitos
Los conjuntos infinitos pueden ser contables o no contables. Los conjuntos contables tienen una correspondencia uno a uno con los números naturales, mientras que los conjuntos no contables no tienen tal correspondencia.

| Conjunto | Tipo | Ejemplo |
| -------- | ---- | ------- |
| Conjunto de Números Naturales (N) | Contable | {0, 1, 2, 3, ...} |
| Conjunto de Números Enteros (Z) | Contable | {..., -2, -1, 0, 1, 2, ...} |
| Conjunto de Números Reales (R) | No contable | Todos los números decimales |

### 8. Comparación de Cardinalidades
Para comparar cardinalidades de conjuntos infinitos, se utiliza la noción de correspondencia uno a uno:

- Contable: Existen correspondencias uno a uno con los números naturales.
- No contable: No existe tal correspondencia, por lo que el conjunto es más grande.
