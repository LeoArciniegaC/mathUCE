##I. Algoritmos:

Es un conjunto ordenado de pasos finitos para solucionar un problema. Aquí tenemos dos clasificaciones:

#####a) Algoritmos no computacionales:

Son, por ejemplo, los que vivimos a diario, los pasos que tenemos para realizar alguna actividad, estos los solemos ejecutar de manera inconsciente.

#####b) Algoritmos computacionales:

Primero hay que entender que una computadora es un conjunto de datos binarios (1 y 0) que necesita una instrucción para poder ejecutar alguna acción, entonces los algoritmos computacionales son instrucciones en forma de código que se dicta a un ordenador o computadora para ejecutar un programa específico.

###1. Estructura básica de un algoritmo:

- En primer lugar, un algoritmo siempre tendrá un inicio y un fin.
- Tiene una serie _**lógica y ordenada**_ de instrucciones a seguir.
- Estas instrucciones inician con un verbo en infinitivo.
- Las instrucciones estarán numeradas.
- Generalmente, lo primero que se hace es recolectar información para luego procesarla y finalmente mostrar un resultado.

###2. Fases de un algoritmo (EPS):

Los algoritmos tienen tres fases principales, descritas por la estructura Entrada, Proceso y Salida (EPS):

####a) Entrada:

La acción que realiza es leer datos que el usuario va a ingresar, según lo requiera el programa.

####b) Proceso:

La acción a realizar es un cálculo o proceso interno a partir de los datos ingresados en la fase anterior.

####c) Salida:

La acción a realizar es la escritura de los resultados obtenidos con los procesos internos que el programa realizó en la fase anterior.

##### Esto podemos observarlo fácilmente en un código sencillo de Python:



```python
#Primero, le preguntamos al usuario la información necesaria para ejecutar el programa, en este caso, una resta:
print("Ingrese el minuendo de la resta a realizar")
min=float(input()) #Este código (input) es para permitir al usuario ingresar la información solicitada, en este caso, el valor del minuendo
print("Ingrese el sustraendo de la resta a realizar")
sustr=float(input()) #Nuevamente, esta línea es para que el usuario ingrese el dato solicitado
#Aquí finalizaría la fase de ENTRADA

#Esta es la fase de PROCESO:
dif=min-sustr #Esta línea ejecuta el cálculo de la resta, a partir de los datos ingresados anteriormente

#Esta es la fase de SALIDA:
print("El resultado de la resta ", min, "-", sustr, " es: ", dif) #Esta línea le muestra al usuario el resultado del proceso realizado internamente por el programa
```

##II. Algoritmos de Búsqueda

###1. Búsqueda Lineal:

Es el método más sencillo hablando de complejidad en los procesos, pues revisa cada elemento del conjunto uno por uno hasta hallar el elemento objetivo.

#####Por ejemplo, supongamos que tenemos los siguientes datos: (3, 8, 7, 16, 45), y se pretende hallar el número 16.

- Se empieza analizando el primer elemento (3), y como no es el número buscado (16), se continúa al siguiente número.
- El siguiente es 8, (no es 16).
- 7 (no es 16).
- El siguiente elemento es el número objetivo, así que termina el algoritmo.

En el ejemplo anterior, el algoritmo tuvo que analizar el 80% de los datos hasta llegar al número objetivo. En este caso, el conjunto era de 5 datos, pero si se aumenta el número de datos del conjunto, aumenta el tiempo que tomaría analizar todos los datos, es decir, aumenta el número de veces que el algoritmo tiene que ejecutarse.

###2. Búsqueda Binaria:

Se emplea en listas de datos ordenados, y es mucho más eficiente y rápido. Su mecanismo es descartar la mitad de los datos en cada paso.

#####Supongamos que tenemos datos de números desde el 1 hasta el 20 = [1 , 20], y pretendemos hallar el número 13.

- Primero preguntamos si el número es mayor o menor a 10 (la mitad del conjunto de datos), en este caso, es mayor, así que ya descartamos el 50% de los datos.
- Luego hallamos la nueva mitad y preguntamos si el número es mayor o menor a 15, en este caso es menor, y ya nos quedamos con el 25% de los datos iniciales.
-Nuevamente hallamos la nueva mitad y preguntamos si el número es mayor o menor a 13, y ese era el número a hallar, culminando la búsqueda.

En el ejemplo anterior, la búsqueda lineal habría demandado 13 pasos hasta encontrar el número buscado, pero con la búsqueda binaria ese proceso se redujo a apenas 3 pasos.

##III. Sistemas Numéricos:

Son formas de representar números. Se conforman de $n$ elementos ordenados, que aumentan de $n$ en $n$ [$n$, $n+1$, $n+2$, ...].

Un elemento va a tener dos clases de valores:

###a)Valor individual:

Es el que está dado según su sistema.

Por ejemplo, en el sistema decimal, el 8 representa la cantidad de ocho.

###b)Valor por su posición:

Depende de la posición que ocupa en el número representado.

Por ejemplo, en $8$, el 8 representa ocho unidades.

En $80$, el 8 representa ocho decenas, ya no unidades.
