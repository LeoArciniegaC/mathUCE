##I. Lógica Proposicional:

###1. Conectores lógicos:

a) Conjunción ("y"): ∧

b) Disyunción ("o"): ∨

c) Implicación ("si, entonces"): →

d) Doble implicación ("sí y solo si"): ↔

e) Negación ("no"): ¬

####1.1. Cómo aplicar los conectores lógicos en Python:

A continuación, entre comillas, los comandos de Python equivalentes a los conectores lógicos:

a) Conjunción (∧) = "and"

b) Disyunción (∨) = "or"

c) Negación (¬) = "not"

###2. Tablas de Verdad y Ejemplos de Conectores Lógicos:

#####a) Conjunción:

| P | Q | P ∧ Q |
|---|---|-------|
| V | V |   V   |
| V | F |   F   |
| F | V |   F   |
| F | F |   F   |

Será verdadero siempre y cuando los dos elementos sean verdaderos, como se aprecia en la tabla.

#####b) Disyunción

| P | Q | P ∨ Q |
|---|---|-------|
| V | V |   V   |
| V | F |   V   |
| F | V |   V   |
| F | F |   F   |

Será verdadero siempre y cuando uno de los dos elementos sea verdadero, como se aprecia en la tabla.

#####c) Implicación:

| P | Q | P → Q |
|---|---|-------|
| V | V |   V   |
| V | F |   F   |
| F | V |   V   |
| F | F |   V   |

Solamente será falso cuando el primer elemento sea verdadero y el segundo sea falso, en el resto de casos será verdadero.

#####d) Doble implicación:

| P | Q | P ↔ Q |
|---|---|-------|
| V | V |   V   |
| V | F |   F   |
| F | V |   F   |
| F | F |   V   |

Será verdadero cuando ambos elementos sean iguales, es decir, cuando ambos sean verdaderos o ambos sean falsos, como se muestra en la tabla.

#####e) Negación

| P | ¬P |
|---|----|
| V | F  |
| F | V  |

La negación es el contrario del elemento, es decir, será verdadera cuando el elemento sea falso, y será falsa cuando el elemento sea verdadero.

###3. Cuantificadores Lógicos:

a) Existencial (∃): "Existe"

b) Universal (∀): "Para todo"

#####Ejemplos:

- ∃x ∀y: "Existe un x para todo y"

- ∀y ∃x: "Para todo y existe un x"

######Nota: Como se puede notar, el orden de los cuantificadores sí es importante y altera el resultado final.

##II. Conjuntos Numéricos:

###1. Conjuntos numéricos básicos y su notación:

- a) Naturales (ℕ): {0, 1, 2, 3, ...}, es decir, son todos los números enteros con signo positivo.

- b) Enteros (ℤ): {..., -3, -2, -1, 0, 1, 2, 3, ...}, es decir, son todos los número enteros positivos y negativos.

- c) Racionales (ℚ): Incluye a los números enteros además de los números decimales exactos, tanto positivos como negativos.

- c.a) Irracionales (ℝ-ℚ): Son todos los números irracionales inexactos, es decir, cuyo número de decimales es infinito (por ejemplo, π).

- d) Reales (ℝ): Es la agrupación de todos los anteriores, es decir, todos los números existentes.

####Relación de contención:

ℕ ⊂ ℤ ⊂ ℚ ⊂ ℝ

⊂ significa "contenido en"

∈ significa "pertenece"

∉ significa "no pertenece"

Un conjunto está definido entre llaves {}

El conjunto vacío se denota con {∅}

###2. Operaciones con conjuntos:

a) Unión: (A ∪ B), es incluir los elementos de ambos conjuntos.

b) Intersección: (A ∩ B), significa los elementos en común que comparten ambos conjuntos.

c) Diferencia: (A \ B), es los elementos del primer conjunto sustrayendo los elementos que comparte con el segundo conjunto.

d) Complemento_ ($A^c$), teniendo en cuenta que A esté contenido en otro conjunto más grande, su complemento será todos los valores de ese conjunto más grande que no están incluidos en el conjunto A.
