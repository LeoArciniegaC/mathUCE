##I. Funciones Biyectivas: Conceptos Básicos, Clasificaciones y Ejemplos

###1. Introducción:
En matemáticas, una función biyectiva es una función que es tanto inyectiva como suprayectiva. Esto significa que cada elemento del dominio se asigna a un elemento único del codominio, y cada elemento del codominio es la imagen de un único elemento del dominio. Las funciones biyectivas son importantes porque establecen una correspondencia uno a uno entre los elementos del dominio y del codominio.

###2. Conceptos Básicos:
Una función $ f: X \to Y $ es biyectiva si cumple con las siguientes propiedades:

1. **Inyectividad (Función Inyectiva):** Una función $ f $ es inyectiva si $ f(a) = f(b) $ implica $ a = b $. Es decir, diferentes elementos del dominio se mapean a diferentes elementos del codominio.
   
2. **Suprayectividad (Función Suprayectiva):** Una función $ f $ es suprayectiva si para todo $ y $ en el codominio $ Y $, existe al menos un $ x $ en el dominio $ X $ tal que $ f(x) = y $. Es decir, cada elemento del codominio tiene una preimagen en el dominio.

###3. Notación:
- $ f: X \to Y $: $ f $ es una función de $ X $ a $ Y $.
- $ f(x) = y $: $ y $ es la imagen de $ x $ bajo la función $ f $.
- $ \forall $: Para todo (cuantificador universal).
- $ \exists $: Existe (cuantificador existencial).

####Ejemplos de Funciones Biyectivas:

1. **Función Lineal Biyectiva:**
   Sea $ f: \mathbb{R} \to \mathbb{R} $ definida por $ f(x) = 2x + 3 $. Demostraremos que $ f $ es biyectiva.

   - Inyectividad: Supongamos $ f(a) = f(b) $. Entonces, $ 2a + 3 = 2b + 3 $. Restando 3 de ambos lados obtenemos $ 2a = 2b $, y dividiendo por 2 obtenemos $ a = b $. Por lo tanto, $ f $ es inyectiva.

   - Suprayectividad: Sea $ y $ un elemento cualquiera de $ \mathbb{R} $. Debemos encontrar un $ x $ tal que $ f(x) = y $. Resolviendo la ecuación $ 2x + 3 = y $, obtenemos $ x = \frac{y - 3}{2} $, que pertenece a $ \mathbb{R} $. Por lo tanto, $ f $ es suprayectiva.

   Concluimos que $ f(x) = 2x + 3 $ es biyectiva.

2. **Función Exponencial Biyectiva:**
   Sea $ f: \mathbb{R} \to \mathbb{R}^+ $ definida por $ f(x) = e^x $, donde $ \mathbb{R}^+ $ es el conjunto de todos los números reales positivos. Demostraremos que $ f $ es biyectiva.

   - Inyectividad: Supongamos $ f(a) = f(b) $. Entonces, $ e^a = e^b $. Tomando el logaritmo natural de ambos lados obtenemos $ a = b $. Por lo tanto, $ f $ es inyectiva.

   - Suprayectividad: Sea $ y $ un elemento cualquiera de $ \mathbb{R}^+ $. Debemos encontrar un $ x $ tal que $ f(x) = y $. Resolviendo la ecuación $ e^x = y $, obtenemos $ x = \ln(y) $, que pertenece a $ \mathbb{R} $. Por lo tanto, $ f $ es suprayectiva.

   Concluimos que $ f(x) = e^x $ es biyectiva.

##II. Funciones Compuestas: Conceptos Básicos, Clasificaciones y Ejemplos

###1. Introducción:
En matemáticas, una función compuesta es el resultado de aplicar una función a los resultados de otra función. La composición de funciones es una operación que toma dos funciones $ f $ y $ g $ y produce una nueva función $ h $ tal que $ h(x) = f(g(x)) $. La notación de una función compuesta es $ (f \circ g)(x) $, que se lee como "f compuesto con g de x".

###2. Conceptos Básicos:
Sea $ f: A \to B $ y $ g: B \to C $ dos funciones. La función compuesta $ f \circ g $ se define como:
$ (f \circ g)(x) = f(g(x)) $ para todo $ x $ en el dominio de $ g $ tal que $ g(x) $ está en el dominio de $ f $.

###3. Clasificaciones de Funciones Compuestas:
Las funciones compuestas pueden clasificarse de diversas maneras según el tipo y las propiedades de las funciones involucradas.

1. **Composición de Funciones Lineales:**
   Si $ f $ y $ g $ son funciones lineales, entonces $ f \circ g $ también es una función lineal.
   Ejemplo: Si $ f(x) = 2x + 3 $ y $ g(x) = x - 1 $, entonces $ (f \circ g)(x) = f(g(x)) = f(x - 1) = 2(x - 1) + 3 = 2x - 2 + 3 = 2x + 1 $.

2. **Composición de Funciones Cuadráticas:**
   Si $ f $ y $ g $ son funciones cuadráticas, entonces $ f \circ g $ puede no ser una función cuadrática.
   Ejemplo: Si $ f(x) = x^2 $ y $ g(x) = 2x + 1 $, entonces $ (f \circ g)(x) = f(g(x)) = f(2x + 1) = (2x + 1)^2 = 4x^2 + 4x + 1 $.

3. **Composición de Funciones Trigonométricas:**
   La composición de funciones trigonométricas produce una función compuesta que puede tener propiedades específicas de trigonometría.
   Ejemplo: Si $ f(x) = \sin(x) $ y $ g(x) = \cos(x) $, entonces $ (f \circ g)(x) = f(g(x)) = \sin(\cos(x)) $.

###4. Notación Matemática:
- $ f \circ g $: Función compuesta de $ f $ y $ g $.
- $ (f \circ g)(x) $: Evaluación de la función compuesta en $ x $.
- $ \forall $: Para todo (cuantificador universal).
- $ \exists $: Existe (cuantificador existencial).

####Ejemplos de Funciones Compuestas:

1. **Función Compuesta de Funciones Lineales:**
   Sea $ f(x) = 3x + 2 $ y $ g(x) = x - 4 $. Calculamos $ (f \circ g)(x) $ y $ (g \circ f)(x) $.

   - $ (f \circ g)(x) = f(g(x)) = f(x - 4) = 3(x - 4) + 2 = 3x - 12 + 2 = 3x - 10 $.
   - $ (g \circ f)(x) = g(f(x)) = g(3x + 2) = (3x + 2) - 4 = 3x - 2 $.

2. **Función Compuesta de Función Lineal y Cuadrática:**
   Sea $ f(x) = x^2 + 1 $ y $ g(x) = 2x $. Calculamos $ (f \circ g)(x) $ y $ (g \circ f)(x) $.

   - $ (f \circ g)(x) = f(g(x)) = f(2x) = (2x)^2 + 1 = 4x^2 + 1 $.
   - $ (g \circ f)(x) = g(f(x)) = g(x^2 + 1) = 2(x^2 + 1) = 2x^2 + 2 $.

3. **Función Compuesta de Funciones Trigonométricas:**
   Sea $ f(x) = \sin(x) $ y $ g(x) = \cos(x) $. Calculamos $ (f \circ g)(x) $ y $ (g \circ f)(x) $.

   - $ (f \circ g)(x) = f(g(x)) = \sin(\cos(x)) $.
   - $ (g \circ f)(x) = g(f(x)) = \cos(\sin(x)) $.

##III. Funciones Inversas: Conceptos Básicos, Clasificaciones y Ejemplos

###1. Introducción:
En matemáticas, una función inversa es una función que "deshace" la acción de la función original. Si $ f $ es una función con dominio $ X $ y codominio $ Y $, entonces su inversa $ f^{-1} $ es una función con dominio $ Y $ y codominio $ X $ tal que $ f^{-1}(f(x)) = x $ para todo $ x $ en $ X $, y $ f(f^{-1}(y)) = y $ para todo $ y $ en $ Y $.

###2. Conceptos Básicos:
Para que una función $ f: X \to Y $ tenga una inversa, debe ser biyectiva, es decir, tanto inyectiva como suprayectiva.

1. **Inyectividad:** Una función $ f $ es inyectiva si $ f(a) = f(b) $ implica $ a = b $.
2. **Suprayectividad:** Una función $ f $ es suprayectiva si para todo $ y $ en el codominio $ Y $, existe al menos un $ x $ en el dominio $ X $ tal que $ f(x) = y $.

###3. Notación Matemática:
- $ f^{-1} $: Función inversa de $ f $.
- $ f: X \to Y $: $ f $ es una función de $ X $ a $ Y $.
- $ f(x) = y $: $ y $ es la imagen de $ x $ bajo la función $ f $.
- $ f^{-1}(y) = x $: $ x $ es la preimagen de $ y $ bajo la función $ f^{-1} $.
- $ \forall $: Para todo (cuantificador universal).
- $ \exists $: Existe (cuantificador existencial).

####Ejemplos de Funciones Inversas:

1. **Función Lineal:**
   Sea $ f(x) = 2x + 3 $. Para encontrar la inversa $ f^{-1}(x) $:
   - Primero, igualamos $ y = 2x + 3 $ y resolvemos para $ x $:
     $$
     y = 2x + 3 \\
     y - 3 = 2x \\
     x = \frac{y - 3}{2}
     $$
   - Por lo tanto, la inversa es $ f^{-1}(x) = \frac{x - 3}{2} $.
   - Verificación: $ f(f^{-1}(x)) = 2\left(\frac{x - 3}{2}\right) + 3 = x - 3 + 3 = x $ y $ f^{-1}(f(x)) = \frac{2x + 3 - 3}{2} = x $.

2. **Función Cuadrática:**
   La función cuadrática $ f(x) = x^2 $ no tiene inversa en todo su dominio porque no es inyectiva. Sin embargo, restringiendo el dominio a $ x \geq 0 $, podemos encontrar la inversa.
   - Sea $ f(x) = x^2 $ para $ x \geq 0 $. Para encontrar la inversa $ f^{-1}(x) $:
     $$
     y = x^2 \\
     x = \sqrt{y}
     $$
   - Por lo tanto, la inversa es $ f^{-1}(x) = \sqrt{x} $ para $ x \geq 0 $.
   - Verificación: $ f(f^{-1}(x)) = (\sqrt{x})^2 = x $ para $ x \geq 0 $ y $ f^{-1}(f(x)) = \sqrt{x^2} = x $ para $ x \geq 0 $.

3. **Función Exponencial y Logarítmica:**
   Sea $ f(x) = e^x $. La inversa de $ f $ es la función logarítmica natural $ f^{-1}(x) = \ln(x) $.
   - Verificación: $ f(f^{-1}(x)) = e^{\ln(x)} = x $ y $ f^{-1}(f(x)) = \ln(e^x) = x $.


##V. Funciones de Utilidad en Ciencias Económicas: Conceptos Básicos, Clasificaciones y Ejemplos

###1. Introducción:
En ciencias económicas, una función de utilidad representa la satisfacción o preferencia de un consumidor por ciertos bienes o servicios. Estas funciones son fundamentales para analizar el comportamiento del consumidor, la teoría de la elección y la teoría del bienestar. La utilidad se mide en unidades arbitrarias llamadas "utiles".

###2. Conceptos Básicos:
Una función de utilidad $ U: X \to \mathbb{R} $ asigna un número real a cada combinación de bienes en el conjunto de consumo $ X $. La utilidad puede ser cardinal o ordinal.

1. **Utilidad Cardinal:** La utilidad cardinal implica que las diferencias en los valores de la función de utilidad tienen significado. Por ejemplo, una utilidad de 10 es el doble de una utilidad de 5.
2. **Utilidad Ordinal:** La utilidad ordinal implica que solo el orden de las preferencias importa. Por ejemplo, una utilidad de 10 indica una preferencia mayor que una utilidad de 5, pero no se puede decir cuánto mayor.

###3. Clasificaciones de Funciones de Utilidad:

1. **Función de Utilidad Lineal:**
   Una función de utilidad lineal toma la forma $ U(x_1, x_2, \ldots, x_n) = a_1 x_1 + a_2 x_2 + \ldots + a_n x_n $ donde $ a_i $ son constantes.
   - Ejemplo: $ U(x_1, x_2) = 3x_1 + 2x_2 $ indica que el bien 1 es preferido al bien 2.

2. **Función de Utilidad Cuadrática:**
   Una función de utilidad cuadrática incluye términos cuadráticos y de interacción entre bienes.
   - Ejemplo: $ U(x_1, x_2) = x_1^2 + 2x_1 x_2 + x_2^2 $.

3. **Función de Utilidad Cobb-Douglas:**
   Una función de utilidad Cobb-Douglas toma la forma $ U(x_1, x_2) = x_1^{a} x_2^{b} $ donde $ a $ y $ b $ son constantes.
   - Ejemplo: $ U(x_1, x_2) = x_1^{0.5} x_2^{0.5} $.

4. **Función de Utilidad CES (Elasticidad de Sustitución Constante):**
   Una función de utilidad CES toma la forma $ U(x_1, x_2) = \left( a x_1^{\rho} + b x_2^{\rho} \right)^{1/\rho} $ donde $ \rho \neq 0 $.
   - Ejemplo: $ U(x_1, x_2) = \left( 0.5 x_1^{-1} + 0.5 x_2^{-1} \right)^{-1} $.

###4. Notación Matemática:
- $ U(x_1, x_2, \ldots, x_n) $: Función de utilidad que depende de $ n $ bienes.
- $ \frac{\partial U}{\partial x_i} $: Derivada parcial de $ U $ con respecto al bien $ x_i $.
- $ MRS_{ij} = -\frac{\frac{\partial U}{\partial x_i}}{\frac{\partial U}{\partial x_j}} $: Tasa Marginal de Sustitución entre los bienes $ x_i $ y $ x_j $.

####Ejemplos de Funciones de Utilidad:

1. **Función de Utilidad Lineal:**
   Sea $ U(x_1, x_2) = 3x_1 + 2x_2 $.
   - Para $ x_1 = 2 $ y $ x_2 = 3 $: $ U(2, 3) = 3(2) + 2(3) = 6 + 6 = 12 $.

2. **Función de Utilidad Cobb-Douglas:**
   Sea $ U(x_1, x_2) = x_1^{0.5} x_2^{0.5} $.
   - Para $ x_1 = 4 $ y $ x_2 = 1 $: $ U(4, 1) = 4^{0.5} \cdot 1^{0.5} = 2 \cdot 1 = 2 $.

3. **Función de Utilidad CES:**
   Sea $ U(x_1, x_2) = \left( 0.5 x_1^{-1} + 0.5 x_2^{-1} \right)^{-1} $.
   - Para $ x_1 = 2 $ y $ x_2 = 2 $: $ U(2, 2) = \left( 0.5 \cdot 2^{-1} + 0.5 \cdot 2^{-1} \right)^{-1} = \left( 0.25 + 0.25 \right)^{-1} = 2 $.
