##I. Función:

Es la relación entre dos conjuntos ${A}$ y $B$, donde $A$ es el dominio (conjunto de salida) y $B$ es su imagen, rango o recorrido (conjunto de llegada).

Para que esta correspondencia sea una función, a los elementos del conjunto de salida les corresponde uno y solo un elemento del conjunto de llegada.

Para escribir esta relación, se usa la expresión:

$f: A→B$

$x→ f(x)$

$f(x) = y$

###1. Tipos de funciones:
####a) Inyectiva:
A cada elemento del conjunto de salida le corresponde un único elemento del conjunto de llegada, y viceversa.
####b) Sobreyectiva:
Todos los elementos del conjunto de llegada son correspondidos por todos los elementos del conjunto de salida, y viceversa.
####c) Biyectiva:
Una función es biyectiva cuando es inyectiva y sobreyectiva a la vez.

##II. Imagen y Preimagen

###1. Conceptos Básicos

En matemáticas, los conceptos de imagen y preimagen son fundamentales en el estudio de funciones. Entender estos conceptos es crucial para el análisis y la manipulación de funciones en diversas áreas, incluyendo álgebra, cálculo y teoría de conjuntos.

###2. Imagen de una Función

La **imagen** de una función $f: X → Y$ es el conjunto de todos los elementos de $Y$ que tienen al menos un predecesor en $X$. En otras palabras, es el conjunto de todos los valores que $f$ toma.

####a)  Notación

Si $f: X → Y$ es una función y $A$ es un subconjunto de $X$, la imagen de $A$ bajo $f$, denotada como $f(A)$, se define como:

$f(A) = \{ f(x) \mid x \in A \}$

La imagen de toda la función $f$ se denota como $Im(f)$ o $f(X)$:

$\text{Im}(f) = \{ f(x) \mid x \in X \}$

##### Ejemplo

Considera la función $f: \mathbb{R} → \mathbb{R}$ definida por $f(x) = x^2$

- Para el conjunto A = {-2, -1, 0, 1, 2}, la imagen $f(A)$ es {0, 1, 4}.
- La imagen de la función $f$ sobre todos los números reales, $Im(f)$, es $(\mathbb{R}_{\geq 0}$, es decir, todos los números reales no negativos.

###3. Preimagen de una Función

La **preimagen** (o contrarrecíproco) de una función ·f: X → Y· es el conjunto de todos los elementos de $X$ que son mapeados a elementos de un subconjunto $B$ de $Y$. En otras palabras, es el conjunto de todos los valores de $X$ que, al ser aplicados a $f$, resultan en elementos de $B$.

####a) Notación

Si $f: X → Y$ es una función y $B$ es un subconjunto de $Y$, la preimagen de $B$ bajo $f$, denotada como $f^{-1}(B)$, se define como:

$f^{-1}(B) = \{ x \in X \mid f(x) \in B \}$

##### Ejemplo

Considera la misma función $f: \mathbb{R} → \mathbb{R}$ definida por $f(x) = x^2$.

- Para el conjunto $B$ = \{0, 1, 4\}, la preimagen $f^{-1}(B)$ es \{0, 1, -1, 2, -2\}.
- Para el conjunto $B$ = \{4\}, la preimagen $f^{-1}(B)$ es \{2, -2\}.

###4. Propiedades Importantes

1. **Inyección y Preimagen:**
   - Si $f$ es inyectiva (una función inyectiva o uno-a-uno), entonces cada elemento de $Y$ tiene como máximo una preimagen en $X$.

2. **Sujeción y Imagen:**
   - Si $f$ es suprayectiva (una función sobreyectiva o sobre), entonces la imagen de $f$ es todo el conjunto $Y$, es decir, $Im(f) = Y$.

3. **Biyectividad:**
   - Si $f$ es biyectiva (tanto inyectiva como suprayectiva), entonces cada elemento de $Y$ tiene exactamente una preimagen en $X$, y la imagen de $f$ es todo $Y$.

4. **Preimagen del Complemento:**
   - La preimagen del complemento de un conjunto es el complemento de la preimagen del conjunto:
   $f^{-1}(Y \setminus B) = X \setminus f^{-1}(B)$

5. **Imagen de la Unión:**
   - La imagen de la unión de dos conjuntos es la unión de las imágenes de los conjuntos:
   $f(A \cup B) = f(A) \cup f(B)$






##III. Funciones Matemáticas: Conceptos Básicos, Clasificaciones y Ejemplos

###1. Introducción:
En matemáticas, una función es una relación entre un conjunto de entradas (dominio) y un conjunto de salidas (codominio), donde cada entrada tiene asignada exactamente una salida. Las funciones son fundamentales en matemáticas y se utilizan para modelar situaciones del mundo real, así como para realizar cálculos y análisis.

###2. Conceptos Básicos:
Una función f se denota usualmente como:

$f: X→Y$

donde X es el dominio de la función (conjunto de todas las posibles entradas) y Y es el codominio (conjunto de todas las posibles salidas). La notación $f(x)$ representa la salida correspondiente a la entrada x en la función f.

###3. Clasificaciones de Funciones:
Las funciones pueden clasificarse de diversas maneras:

1. Según el tipo de variables:

- Funciones univariadas: Tienen una única variable independiente.
  Ejemplo: $f(x) = 2x + 1$

- Funciones multivariadas: Tienen más de una variable independiente.
  Ejemplo: $f(x, y) = x^2 + y^2$

2. Según la relación entre las variables:

- Funciones lineales: Tienen la forma $f(x) = ax + b$, donde a y b son constantes.
  Ejemplo: $f(x) = 3x + 2$

- Funciones cuadráticas: Tienen la forma $f(x) = ax^2 + bx + c$, donde a, b, y c son constantes y a ≠ 0.
  Ejemplo: $f(x) = x^2 + 4x + 3$

- Funciones exponenciales: Tienen la forma $f(x) = a^x$, donde a es una constante.
  Ejemplo: $f(x) = 2^x$

- Funciones trigonométricas: Incluyen funciones como seno $sin(x)$, coseno $cos(x)$, tangente $tan(x)$, etc.

3. Según el comportamiento de la función:

- Funciones crecientes y decrecientes: Dependiendo de cómo cambian sus valores a medida que aumenta la variable independiente.
  
- Funciones periódicas: Aquellas que se repiten a intervalos regulares.

Ejemplos de Funciones:

1. Función Lineal:
   $f(x) = 3x + 1$
   - Para $x = 2: f(2) = 7$

2. Función Cuadrática:
   $f(x) = x^2 - 4x + 4$
   - Para $x = 3: f(3) = 1$

3. Función Exponencial:
   $f(x) = 2^x$
   - Para $x = 2: f(2) = 4$

4. Función Trigonométrica:
   $f(x) = sin(x)$
   - Para $x = π/2: f(π/2) = 1$




##IV. Principio de Demostración en Matemáticas: Conceptos Básicos, Clasificaciones y Ejemplos

###1. Introducción:
En matemáticas, una demostración es el proceso lógico mediante el cual se establece la veracidad de una afirmación matemática. La demostración es fundamental porque proporciona una justificación rigurosa y verificable de los resultados matemáticos.

###2. Conceptos Básicos:
Una demostración matemática consta típicamente de una serie de pasos lógicos y deductivos que parten de supuestos o axiomas aceptados hasta llegar a una conclusión válida. Se emplean reglas de inferencia y técnicas específicas para asegurar la validez de cada paso.

###3. Clasificaciones de Demostraciones:
Las demostraciones pueden clasificarse de varias maneras:

1. Según el método utilizado:

- Demostraciones directas: Se procede directamente desde los axiomas o supuestos hasta la conclusión deseada mediante una serie de pasos lógicos.
  Ejemplo: Demostrar que la suma de dos números pares siempre es un número par.

- Demostraciones por contraposición: Se demuestra la afirmación contraria y se concluye que la afirmación original es verdadera.
  Ejemplo: Demostrar que si $n^2$ es par, entonces $n$ es par.

- Demostraciones por contradicción: Se asume lo contrario de lo que se quiere probar y se llega a una contradicción lógica, demostrando así la veracidad de la afirmación original.
  Ejemplo: Demostrar que no existen números racionales cuyos cuadrados sean 2.

2. Según el área de aplicación:

- Demostraciones en álgebra: Usualmente se emplean técnicas algebraicas para manipular ecuaciones y expresiones.
  
- Demostraciones en geometría: Se utilizan propiedades geométricas y teoremas específicos para establecer relaciones y resultados.

###4. Notaciones Matemáticas Comunes:
- $\Rightarrow $: Implicación lógica.
- $\Leftrightarrow$: Equivalencia lógica.
- $ \therefore $ : Por lo tanto, utilizado para denotar conclusión.
- $ \because $ : Porque, utilizado para denotar razón.

####Ejemplos de Demostraciones:

1. Demostración Directa:
   Para demostrar que la suma de dos números pares siempre es un número par:
   Sea $ a $ y $ b $ dos números pares. Entonces, existen enteros $ m $ y $ n $ tales que $ a = 2m $ y $ b = 2n $. Por lo tanto, $ a + b = 2m + 2n = 2(m + n) \$, que es par.

2. Demostración por Contraposición:
   Para demostrar que si $ n^2 $ es par, entonces $ n $ es par:
   Supongamos que $ n $ es impar. Entonces $ n = 2k + 1 $ para algún entero $ k $. Entonces $ n^2 = (2k + 1)^2 = 4k^2 + 4k + 1 $, que es impar. Por lo tanto, si $ n^2 $ es par, entonces $ n $ debe ser par.

3. Demostración por Contradicción:
   Para demostrar que no existen números racionales cuyos cuadrados sean 2:
   Supongamos que existe un número racional $ \frac{p}{q} $ tal que $ \left(\frac{p}{q}\right)^2 = 2 $. Entonces $ p^2 = 2q^2 $. Como $ p^2 $ es par, $ p $ debe ser par. Si $ p $ es par, $ p^2 $ es divisible para 4. Sin embargo, $ 2q^2 $ no es divisible para 4, lo cual lleva a una contradicción. Por lo tanto, no existen números racionales cuyos cuadrados sean 2.
